## Questão 2

```python
# Carrega as métricas desejadas
from sklearn.metrics import accuracy_score, roc_auc_score

# Data
X = dfTrain.drop(['Class'], axis=1)
# Target
y = dfTrain['Class']
```

### 2a)

```python
# Número de repetições para teste com o vetor randômico com metade de 1s e
# metade de 0s
N_RUNS = 30
CLASS_TYPES = [
    ['50/50', 0, 0],
    ['100%', 0, 0],
    ['0%', 0, 0],
]


artificial: np.ndarray = np.array([])

# Para estimador com 100% de fraudes
artificial = np.ones(y.shape)
CLASS_TYPES[1][1] = accuracy_score(y, artificial)
CLASS_TYPES[1][2] = roc_auc_score(y, artificial)

# Para estimador com 0% de fraudes
artificial = np.zeros(y.shape)
CLASS_TYPES[2][1] = accuracy_score(y, artificial)
CLASS_TYPES[2][2] = roc_auc_score(y, artificial)

# Para estimador com 50% de fraudes
for _ in range(N_RUNS):
    artificial = np.random.choice(a=[0,1], size=y.shape)
    CLASS_TYPES[0][1] += accuracy_score(y, artificial)/N_RUNS
    CLASS_TYPES[0][2] += roc_auc_score(y, artificial)/N_RUNS

for _TYPE in CLASS_TYPES:
    print(
        "Modelo {}:\n".format(_TYPE[0]),
        "\tAcurácia: {}\n".format(_TYPE[1]),
        "\tAUC: {}".format(_TYPE[2]),
        "\n"
    )
```

### 2b)

Como podemos ver pelos dados obtidos, a métrica a ser adotada deve ser o AUC, já que o número de entradas da classe 1 é ordens de grandeza menor que as de class 0, assim um modelo que simplesmente classifica todas as entradas como 0 atinge uma acurácia de > 99%, enquanto o _score_ AUC se mantém estável para todos os conjuntos artificialmente gerados.

## Questão 3

```python
# Imports necessários
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from typing import Dict, List
```

### 3a)

```python
# Conjunto de parâmetros para teste
paramGrid: Dict[str, List[int]] = {
    'n_estimators': [ 10, 50, 100, 200 ],
    'max_depth': [ 2, 3, 4, 5 ],
    'random_state': [ 42 ],
}

# Testa conjunto
rf = RandomForestClassifier()
clf = GridSearchCV(rf, paramGrid, cv=3, scoring='roc_auc')
clf.fit(X, y)
```

Temos como os melhores parâmetros para o modelo então:

```python
clf.best_params_
```

Com esses parâmetros, temos um score de:

```python
clf.best_score_
```

### 3b)

Esse parâmetro regula os 'pesos' no cálculo da árvore de decisão referentes a cada classe. Quando utilizado o hiper parâmetro como 'balanced', os pesos são regulados de forma inversamente proporcional a sua frequência na base de dados. Assim, maior peso de decisão é dado a classe com menor representação dentre os dados, permitindo que considere melhor as poucas entradas de classe '1'.

```python
paramGrid2: Dict[str, List[int]] = {
    'n_estimators': [ 10, 50, 100, 200 ],
    'max_depth': [ 2, 3, 4, 5 ],
    'random_state': [ 42 ],
    'class_weight': [ 'balanced', ]
}

rf2 = RandomForestClassifier()
clf2 = GridSearchCV(rf2, paramGrid2, cv=3, scoring='roc_auc')
clf2.fit(X, y)
```

Temos como os melhores parâmetros para o modelo com classes balanceadas então:

```python
clf2.best_params_
```

Com esses parâmetros, temos um score de:

```python
clf2.best_score_
```

Podemos ver que ao balancear os pesos das classes o número de melhor número de estimadores diminui, já que assim os estimadores são melhores escolhas, e mesmo assim obtemos um moodelo com melhor AUC. Esse será o modelo então selecionado para a questão seguinte.

### Q4a)

```python
# Melhor modelo
rf3 = RandomForestClassifier(
    class_weight='balanced',
    max_depth=5,
    n_estimators=50,
    random_state=42,
)
rf3.fit(X, y)
```

```python
dfTest = pd.read_csv(getPath("test.csv"))
# Retiramos os metadados da tabela
dfTest = dfTest.drop(['Time'], axis=1)
dfTest
```

```python
# Dados
XTest = dfTest.drop(['Class'], axis=1)
# Target
yTest = dfTest['Class']
```

```python
# Tomamos a base de testes para análise
dfTest2 = dfTest
# Predição com base nos dados do dataframe
predicts = rf3.predict_proba(XTest)
# Classe obtida pelo modelo
dfTest2['Predict'] = np.array([ np.argmax(i) for i in predicts ])
# Probabilidade de não ser fraude
dfTest2['ProbTrue'] = np.array([ i[0] for i in predicts ])
# Probabilidade de fraude
dfTest2['ProbFraud'] = np.array([ i[1] for i in predicts ])
dfTest2
```

```python
# Dataframe com os dados ordenados por probabilidade de fraude
dfFraud = dfTest2.sort_values(by=['ProbFraud'], ascending=False)
dfFraud
```

```python
# Top 1% entradas por probabilidade de fraude
dfFraud = dfFraud.head(dfFraud.shape[0] // 100)
dfFraud
```

```python
# Lucro dadas as regras
lucro = len(dfFraud[(dfFraud['Class'] == 1)])*100 - \
    len(dfFraud[(dfFraud['Class'] == 0)])*2

print("Lucro trazido pelo modelo: R${:.2f}".format(lucro*1.))
```
